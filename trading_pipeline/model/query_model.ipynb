```python
bullish_prompts = pd.read_csv("../prompts/wo_portfolio_prompts_dfs/bullish_prompts.csv", parse_dates=['timestamp'])
bullish_prompts.head(1)
```

```python
# For testing
trial = bullish_prompts.iloc[0]["prompt"]
print(trial)
```

```python
# API KEY
client = OpenAI(api_key="TODO")

# Function to query the model with a prompt
# To be used in the automation pipeline
def query_model(prompt: str, model: str = "gpt-4o", temperature: float = 0.3) -> str:
    response = client.chat.completions.create(
        model = model,
        messages = [
            {"role": "user", "content": prompt}
        ],
        temperature=temperature,
    )
    return response.choices[0].message.content.strip()
```

```python
response = query_model(trial)
print(response)
```
