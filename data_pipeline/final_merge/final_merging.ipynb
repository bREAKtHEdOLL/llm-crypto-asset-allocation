```python
import os
import pandas as pd
```

### BTC

```python
btc_more_stats = pd.read_csv("BTC/btc_more_stats.csv", parse_dates=["timestamp"])
btc_unique_wallets = pd.read_csv("BTC/btc_unique_wallets.csv", parse_dates=["timestamp"])
btc_price_volume = pd.read_csv("BTC/btc_price_volume.csv", parse_dates=["timestamp"])
```

```python
btc_more_stats
```

```python
# 补充btc_more_stats的缺失数据
full_hours = pd.date_range(start=btc_more_stats["timestamp"].min(), end=btc_more_stats["timestamp"].max(), freq="H")
missing_hours = full_hours.difference(btc_more_stats["timestamp"])
print(f"总共有 {len(missing_hours)} 个漏掉的小时")
print(missing_hours)
```

```python
btc_more_stats = btc_more_stats.set_index("timestamp")
full_range = pd.date_range(start=btc_more_stats.index.min(), end=btc_more_stats.index.max(), freq="H")
btc_more_stats = btc_more_stats.reindex(full_range)

btc_more_stats = btc_more_stats.reset_index().rename(columns={"index": "timestamp"})
btc_more_stats = btc_more_stats.interpolate(method="linear")

btc_more_stats["transaction_count"] = btc_more_stats["transaction_count"].round(0).astype(int)
btc_more_stats["total_value_transferred_btc"] = btc_more_stats["total_value_transferred_btc"].round(4)
btc_more_stats["avg_fee_btc"] = btc_more_stats["avg_fee_btc"].round(8)
btc_more_stats["total_size_used_mb"] = btc_more_stats["total_size_used_mb"].round(4)
```

```python
# btc_unique_wallets
btc_unique_wallets = btc_unique_wallets[(btc_unique_wallets["timestamp"] >= pd.to_datetime("2024-03-01")) & (btc_unique_wallets["timestamp"] < pd.to_datetime("2025-04-01"))].copy()
btc_unique_wallets
# 有缺失数据
```

```python
btc_unique_wallets = btc_unique_wallets.set_index("timestamp")
full_range = pd.date_range(start=btc_unique_wallets.index.min(), end=btc_unique_wallets.index.max(), freq="H")
btc_unique_wallets = btc_unique_wallets.reindex(full_range)

btc_unique_wallets = btc_unique_wallets.reset_index().rename(columns={"index": "timestamp"})
btc_unique_wallets = btc_unique_wallets.interpolate(method="linear")

btc_unique_wallets["unique_active_wallets"] = btc_unique_wallets["unique_active_wallets"].round(0).astype(int)
btc_unique_wallets
```

```python
# 第一次合并
btc_more_stats_wallets = btc_more_stats.merge(btc_unique_wallets, on="timestamp", how="inner")
btc_more_stats_wallets
```

```python
# 第二次合并
btc_almost = btc_price_volume.merge(btc_more_stats_wallets, on="timestamp", how="inner")
btc_almost
```

```python
btc_almost.to_csv("BTC/btc_almost.csv", index=False)
```

```python
btc_almost.isna().sum()
```

```python
btc_almost
```

```python
# 最后合并market_cap的数据
btc_market_cap = pd.read_csv("BTC/btc_market_cap.csv", parse_dates=["timestamp"])
btc_market_cap
```

```python
btc_final = btc_almost.merge(btc_market_cap, on="timestamp", how="inner")
btc_final
```

```python
# Sanity check
expected = pd.date_range(start="2024-05-01 00:00:00", end="2025-03-31 23:00:00", freq="H")

mismatch_mask = btc_final["timestamp"].values != expected.values
print("有没有不一致的？", mismatch_mask.any())
```

```python
btc_final.to_csv("BTC/btc_final.csv", index=False)
```
